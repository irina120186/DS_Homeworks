```python
import numpy as np
from PIL import Image
from PIL import ImageFilter
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
import requests

url = 'https://i.ibb.co/QCwHvPj/N.jpg'
raw = requests.get(url, stream=True).raw
Img=Image.open(raw)
Img
```

```python
arrImg = np.array(Img)
arrImg.shape
```

```python
arrImg=arrImg.astype('float64')
for i in range(arrImg.shape[0]-2):
    for j in range(arrImg.shape[1]-2):
      #for k in range (arrImg.shape[2]-2):
       for k in range (1, 3):
          arrImg[i][j][k]=(arrImg[i][j][k]+arrImg[i+1][j+1][k]+arrImg[i-1][j-1][k]+arrImg[i+1][j-1][k]
                       +arrImg[i-1][j+1][k]+arrImg[i+1][j][k]+arrImg[i-1][j][k]+arrImg[i][j+1][k]+arrImg[i][j-1][k])/9
arrImg[:,:,0]=arrImg[:,:,0]
arrImg[:,:,1]=arrImg[:,:,1]
arrImg[:,:,2]=arrImg[:,:,2]
pict3=arrImg.astype('uint8')
Image.fromarray(pict3)
```

```python
k=0
for i in range(arrImg.shape[0]-2):
    for j in range(arrImg.shape[2]-2):
        arrImg[i][j][k]=(arrImg[i][j][k]+arrImg[i+1][j+1][k]+arrImg[i-1][j-1][k]+arrImg[i+1][j-1][k]
                       +arrImg[i-1][j+1][k]+arrImg[i+1][j][k]+arrImg[i-1][j][k]+arrImg[i][j+1][k]+arrImg[i][j-1][k])/9

arrImg[:,:,0]
pict3=arrImg.astype('uint8')
Image.fromarray(pict3)
```

```python
for i in range(arrImg.shape[0]):
    for j in range(arrImg.shape[1]):
        for k in range(arrImg.shape[2]):
         
            if arrImg[i][j][k]>130:
                arrImg[i][j][k]=255
            else:
                arrImg[i][j]=0   
gray = arrImg   
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show() 
```
