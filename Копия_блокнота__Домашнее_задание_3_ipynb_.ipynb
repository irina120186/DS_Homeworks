2.1 Сгенерируйте выборку размера 100 из равномерного распределения на отрезке [0,1]

```python
import numpy as np
import matplotlib.pyplot as plt  
from matplotlib import pyplot  as plt
from scipy import stats as sps
from scipy.stats import uniform
```

```python
r = uniform.rvs(size=100)
r
```

```python
rr = sps.uniform(loc=0, scale=1).rvs(size=100)
rr
```

2.2 Нарисуйте на одном графике плотность распределения и гистограмму 

```python
x = np.arange(-1, 2, 0.01)
h = rr
```

```python
z = sps.uniform(loc=0, scale=1).pdf(x)
 
plt.plot(x, z, label='loc:0, sc:1')
 

plt.hist(h, bins=130, density=True)
plt.show()
```
